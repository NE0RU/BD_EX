```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 이미지 업로드

```python
import os
```

```python
train_cats_dir = "./data/cats_and_dogs_filtered/train/cats/"
train_dogs_dir = "./data/cats_and_dogs_filtered/train/dogs/"
test_cats_dir = "./data/cats_and_dogs_filtered/test/cats/"
test_dogs_dir = "./data/cats_and_dogs_filtered/test/dogs/"
```

```python
train_cat_fnames = os.listdir(train_cats_dir)
print("파일이름 :",train_cat_fnames[:3])
print("파일 개수 :",len(train_cat_fnames))

train_dog_fnames = os.listdir(train_dogs_dir)
print("파일이름 :",train_dog_fnames[:3])
print("파일 개수 :",len(train_dog_fnames))

test_cat_fnames = os.listdir(test_cats_dir)
print("파일이름 :",test_cat_fnames[:3])
print("파일 개수 :",len(test_cat_fnames))

test_dog_fnames = os.listdir(test_dogs_dir)
print("파일이름 :",test_dog_fnames[:3])
print("파일 개수 :",len(test_dog_fnames))
```

```python
plt.imread("./data/cats_and_dogs_filtered/train/cats/cat.0.jpg")
```

```python
def load_images(default_path, filenames):
    images = []
    
    for filename in filenames:
        #./data/cats_and_dogs_filtered/train/cats/cat.0.jpg
        filepath = default_path + filename
        
        images.append(plt.imread(filepath))
        
    return np.array(images)
```

```python
X_train_cats = load_images(train_cats_dir, train_cat_fnames)
X_train_cats.shape
```

```python
X_train_dogs = load_images(train_dogs_dir, train_dog_fnames)
X_train_dogs.shape
```

```python
X_test_cats = load_images(test_cats_dir, test_cat_fnames)
X_test_cats.shape
```

```python
X_test_dogs = load_images(test_dogs_dir, test_dog_fnames)
X_test_dogs.shape
```

## 이미지 전처리(사이즈 통일)

```python
# !pip install pillow
```

```python
from PIL import Image
```

```python
np.array(Image.fromarray(X_train_dogs[4]).resize((224,224)))
```

```python
def resize_image(original_images, size):
    resize_images = []
    
    for image in original_images:
        resize_image = np.array(Image.fromarray(image).resize(size))
        
        resize_images.append(resize_image)
        
    return np.array(resize_images)
```

```python
X_train_cats_224 = resize_image(X_train_cats, (224,224))
X_train_cats_224.shape
```

```python
figure, axes = plt.subplots(nrows=1, ncols=5)
figure.set_size_inches(18, 4)

axes[0].imshow(X_train_cats_224[0])
axes[1].imshow(X_train_cats_224[1])
axes[2].imshow(X_train_cats_224[2])
axes[3].imshow(X_train_cats_224[3])
axes[4].imshow(X_train_cats_224[4])
```

```python
X_train_dogs_224 = resize_image(X_train_dogs, (224,224))
X_train_dogs_224.shape
```

```python
X_test_cats_224 = resize_image(X_test_cats, (224,224))
X_test_cats_224.shape
```

```python
X_test_dogs_224 = resize_image(X_test_dogs, (224,224))
X_test_dogs_224.shape
```

## X_train데이터 만들기(강아지,고양이 병합)

```python
X_train = np.concatenate((X_train_cats_224, X_train_dogs_224))
X_train.shape
```

```python
X_test = np.concatenate((X_test_cats_224, X_test_dogs_224))
X_test.shape
```

## y데이터 생성(고양이:0, 강아지:1)

```python
y_train = np.array([0]*1000+[1]*1000)
print(y_train[:10])
print(y_train[-10:])
print(y_train.shape)
```

```python
y_test = np.array([0]*500+[1]*500)
print(y_test[:10])
print(y_test[-10:])
print(y_test.shape)
```

## CNN 모델 생성

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()

model.add(Conv2D(filters=8,
                 kernel_size=(3,3),
                 activation="relu",
                 padding="same",
                 input_shape=(224,224,3)))

model.add(Conv2D(filters=8,
                 kernel_size=(3,3),
                 activation="relu",
                 padding="same"))

model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))


model.add(Conv2D(filters=8,
                 kernel_size=(3,3),
                 activation="relu",
                 padding="same"))

model.add(Conv2D(filters=8,
                 kernel_size=(3,3),
                 activation="relu",
                 padding="same"))

model.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

model.add(Flatten())

model.add(Dense(units=128, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(X_train,#문제
          y_train,#답
          epochs=10,#학습횟수
          validation_data=(X_test, y_test),#검증데이터
          verbose=2)#출력창
```

```python
#평가
model.evaluate(X_test, y_test, verbose=0)
```
