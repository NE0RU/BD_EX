```python
import csv
max_temp = -999
max_date = ''
f = open('test_1980_5.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')

header = next(data)

for row in data :
    if row[-1] == '':
        row[-1] = -999
    row[-1] = float(row[-1])
    if row[-1] > max_temp:
        max_temp = row[-1]
        max_date = row[0]
f.close()

date = max_date
date = date.split('-')
print(date[0] + '년', date[1], '월 광주의 최고 기온이 가장 높았던 날은', date[1], '월', date[2], '일로, ', max_temp, '도 였습니다.')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('test_GwangJu.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')

header = next(data)
high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] != '':
        if 2002 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '01' and row[0].split('-')[2] == '15' :
                high.append(float(row[-1]))
                low.append(float(row[-2]))
f.close()
plt.figure(figsize = (10,2), dpi = 300)

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 깨짐 방지
plt.title('내 생일의 기온 변화 그래프')
plt.plot(high, 'r')
plt.plot(low, 'b')
plt.show()
```
