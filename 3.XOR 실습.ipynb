```python
import numpy as np 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#데이터 생성
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])
```

```python
X
```

```python
y
```

```python
model = Sequential()

model.add(Dense(units=32,
                input_dim=2,
                activation="sigmoid"))

model.add(Dense(units=1,
                activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

```python
model.fit(X,
          y,
          epochs=1000,
          verbose=0)
```

```python
evalu = model.evaluate(X, y, verbose=0)
print("loss :",evalu[0])
print("accuracy :",evalu[1])
```

```python
model.predict(X)
```

```python
model.get_weights()
```

## 다중분류

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[1,0],[0,1],[0,1],[1,0]])
X
```

```python
y
```

```python
model = Sequential()
model.add(Dense(units=32, input_dim = 2, activation="sigmoid"))

model.add(Dense(units=2,
                activation="softmax"))
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="RMSprop",
             metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(X, y, epochs=2000)
```

```python
model.evaluate(X, y, verbose=0)
```

```python
model.predict(X) 
```

```python
model.predict_classes(X)
```
